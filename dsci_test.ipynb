### Kiến thức liên quan:
Trong ô theo sau, học viên định nghĩa các phương thức (hàm) trong lớp Point và Vector nhằm hỗ trợ người dùng: tạo đối tượng và thực hiện các phép toán giữa các đối tượng như mô tả sau:

* Với Point (điểm):
  * Hiệu số hai điểm:
    * Nếu A và B là hai điểm
    * Thì: AB = B - A là véctơ đi từ A đến B
    * Cách thực hiện: AB = [B.x - A.x, B.y - A.y, B.z - A.z]
    
* Với Vector (véctơ, vector):
  * Cộng hai véctơ: 
    * Cách thực hiện: a + b = [a.x + b.x, a.y + b.y, a.z + b.z]
  * Tích vô hướng (dot):
    * Nếu: a = [a.x, a.y, a.z]; b = [b.x, b.y, b.z]
    * Thì: dot(a, b) = a.dot(b) = b.dot(a) = a.x * b.x + a.y * b.y + a.z * b.z
    * Lưu ý: ta cũng có
      * dot(a,b) = length(a) * length(b) * cos(angle(a,b))
        * length: độ lớn của vectơ
        * angle(a, b) là góc tạo bởi hai vectơ
      * do đó, angle(a,b) = arccos(**dot(a,b)/ (length(a) * length(b))**)
  * Tích có hướng (cross): (**code cho sẵn**)
    * Nếu: **a** và **b** là hai véctơ thì tích vô hướng là véctơ **c = a x b**; lưu ý: **a x b <> b x a**
    * Tích có hướng là vectơ (có hướng), được tính thông qua tính định thức của ma trận (cho trong phần code)
    * Lưu ý:
      * Nếu ABC là tam giác thì diện tích của tam giác là S(ABC) = 0.5 * length(**AB** x **AC**)

## D.1. Code mẫu về phân loại các hạt

 *** Chạy các lệnh ở các ô trong phần này để phân loại các hạt***

```python
##########################################################################
# Lưu ý: không chỉnh sửa code trong ô này
# Chỉ thực thi
##########################################################################

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
import time
np.seterr(divide='ignore', invalid='ignore')

def cmatrix(target, prediction, nclasses=None):
    if nclasses is None:
        C = target.max() + 1
    else:
        C = nclasses
    N = target.shape[0]
    M = np.zeros((C, C, N), dtype=np.int64)
    M[target.ravel(), prediction.ravel(), np.arange(N)] = 1
    return M.sum(axis=2)

def precision(target, prediction, nclasses=None):
    M = cmatrix(target, prediction, nclasses)
    if nclasses is None:
        C = target.max() + 1
    else:
        C = nclasses
    weights = M.sum(axis=1)/M.sum()
    prec_per_class = M[np.arange(C), np.arange(C)]/M.sum(axis=0)
    prec_averaged = prec_per_class.mean()
    prec_weighted = (prec_per_class*weights).sum()
    return prec_per_class, prec_averaged, prec_weighted

def recall(target, prediction, nclasses=None):
    M = cmatrix(target, prediction, nclasses)
    if nclasses is None:
        C = target.max() + 1
    else:
        C = nclasses
        
    weights = M.sum(axis=1)/M.sum()
    rec_per_classes =  M[np.arange(C), np.arange(C)]/M.sum(axis=1)
    prec_averaged = rec_per_classes.mean()
    prec_weighted = (rec_per_classes*weights).sum()
    
    return rec_per_classes, prec_averaged, prec_weighted

def f1_measure(target, prediction, nclasses=None):
    prec, _, _ = precision(target, prediction, nclasses)
    rec, _, _ = recall(target, prediction, nclasses)
    M = cmatrix(target, prediction, nclasses)
    if nclasses is None:
        C = target.max() + 1
    else:
        C = nclasses
        
    weights = M.sum(axis=1)/M.sum()
    f1_per_classes =  2*prec*rec/(prec+rec)
    f1_averaged = f1_per_classes.mean()
    f1_weighted = (f1_per_classes*weights).sum()

    return f1_per_classes, f1_averaged, f1_weighted
    
def accuracy(target, prediction, nclasses=None):
    M = cmatrix(target, prediction, nclasses)
    if nclasses is None:
        C = target.max() + 1
    else:
        C = nclasses
    return M[np.arange(C), np.arange(C)].sum()/M.sum()


def my_clf_report(target, prediction, nclasses, digits=2):
    """
        return (
            accuracy_score,
            weighted_avg_precision,
            weighted_avg_precall,
            weighted_avg_f1-score
        )
    """
    
    accuracy_score = accuracy(target, prediction, nclasses)
    _, _, weighted_avg_precision = precision(target, prediction, nclasses)
    _, _, weighted_avg_precall = recall(target, prediction, nclasses)
    _, _, weighted_avg_f1 = f1_measure(target, prediction, nclasses)
    
    return [
        round(accuracy_score, digits),
        round(weighted_avg_precision, digits),
        round(weighted_avg_precall, digits),
        round(weighted_avg_f1, digits)
    ]
def highlight_max(s):
    is_max = s == s.max()
    return ['color: red; font-weight: 700' if v else '' for v in is_max]
```

```python
##########################################################################
# Lưu ý: không chỉnh sửa code trong ô này
# Chỉ thực thi
##########################################################################
def build_models(X_train, X_test, y_train, y_test):
  #Tạo các mô hình - dùng sklearn
  models = {
      'MultinomialNB:': MultinomialNB(),
      'LogisticRegression': LogisticRegression(max_iter=500),
      'KNeighbors': KNeighborsClassifier(n_neighbors=5),
      'DecisionTreeClassifier': DecisionTreeClassifier(max_depth=5),
      'RandomForestClassifier': RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1)
  }
  print('Đã tạo được các mô hình cho các kỹ thuật sau: ')
  print('-'*120)
  print(list(models.keys()))

  #Huấn luyện + đánh giá
  print()
  print('Huấn luyện + Kiểm tra')
  print('-'*120)
  results = {}
  for method in models.keys():
      model = models[method]
      print(method)
      start_time = time.time()
      
      print('\t Huấn luyện ...')
      model.fit(X_train, y_train)
      
      print('\t Kiểm tra ...')
      y_pred = model.predict(X_test)
      elapsed_time = (time.time() - start_time)
      
      #sinh kết quả
      report = my_clf_report(y_test, y_pred, 3, digits=3)
      report.append(elapsed_time)
      results[method] = report
  #
  return results, models

```

```python
##########################################################################
# Lưu ý: không chỉnh sửa code trong ô này
# Chỉ thực thi
##########################################################################

np.random.seed(2021)
variables = ['Area', 'Perimeter', 'Compactness', 'Kernel.Length',	'Kernel.Width',	'Asymmetry.Coeff',	'Kernel.Groove']
X = seed[variables].to_numpy()
y = (seed['Type'].to_numpy() - 1).astype(np.int64)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

results, _ = build_models(X_train, X_test, y_train, y_test)
pd.set_option('display.max_columns', None)
columns = ['[accuracy]', '[weighted precision]', '[weighted recall]', '[weighted f1-score]', '[time]']
df1 = pd.DataFrame.from_dict(results, orient="index", columns=columns)
df1 = df1[columns]

print('Kết quả kiểm tra trên {:f} điểm dữ liệu'.format(X_test.shape[0]) )
print('-'*120)
df1.style.apply(highlight_max).format("{:.2}")
```
